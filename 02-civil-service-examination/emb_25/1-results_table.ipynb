# From the results of a specific test, we get the clean PDF

## For the broader concorrence

```python
import pandas as pd
with open("data/grades_broad_raw.txt", "r", encoding="utf-8") as file:
    test_result = file.read()

# remove \n and split results
test_result = test_result.replace("\n", " ").split("/")

# initialize an dataframe
dataframe = []
dataframe.append(["Name", "General Grade", "General Result",
 "Complementary Grade", "Complementary Result", "Specific grade", "Specific result" ,"Final Grade"])

for indiv in test_result:
    # split individuals 
    indiv = indiv.split(", ")
    complete_name = indiv[1]
    name = complete_name.split(" ")[0]
    p1_grade = indiv[2].removesuffix(".00")
    p1_result = indiv[3]
    p2_grade = indiv[4].removesuffix(".00")
    p2_result = indiv[5]
    p3_grade = indiv[6].removesuffix(".00")
    p3_result = indiv[7]
    final_result = indiv[8].removesuffix(".00 ")
    print(final_result)
 
    dataframe.append([name, p1_grade, p1_result, p2_grade, p2_result, p3_grade, p3_result, final_result])

df = pd.DataFrame(dataframe[1:], columns=dataframe[0])
# save
df.to_csv("results_broad.csv", index=False, encoding="utf-8")
```
